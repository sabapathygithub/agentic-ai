```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from IPython.display import display, Markdown
```

```python
load_dotenv(override=True)
```

```python
agent = Agent(name="FinanceAdvisor", instructions="Provide financial advice to users based on their queries.", model="gpt-4o-mini")
```

```python
with trace("Tell me a finance advice"):
    result = await Runner().run(agent, "What is the best way to save for retirement?")
```
