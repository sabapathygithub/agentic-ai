```python
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
load_dotenv(override=True)
```

```python
ollama_client = OpenAI(base_url="http://localhost:11434/v1",api_key="")
model_name = "llama3.2"
question = "Do you know about Agentic AI?"
messages = [    
    {"role": "user", "content": question}
]

resposne = ollama_client.chat.completions.create(model=model_name, messages=messages)
display(Markdown(resposne.choices[0].message.content))
```
