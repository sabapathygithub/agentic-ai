```python
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
load_dotenv(override=True)
```

```python
question = "Do you know about Agentic AI?"
messages = [    
    {"role": "user", "content": question}
]
```

```python
models = []
answers = []
openai = OpenAI()
model_name = "gpt-4o-mini"
response = openai.chat.completions.create(
    model=model_name, 
    messages=messages
)
models.append(model_name)
answers.append(response.choices[0].message.content)
```

```python
display(Markdown(f"### Model: {model_name}\n\n{response.choices[0].message.content}"))
```

```python
ollama = OpenAI(base_url="http://localhost:11434/v1",api_key="")
model_name = "llama3.2"
response = ollama.chat.completions.create(
    model=model_name, 
    messages=messages
)
models.append(model_name)
answers.append(response.choices[0].message.content)
display(Markdown(f"### Model: {model_name}\n\n{response.choices[0].message.content}"))
```

```python
together = ""
for index, answer in enumerate(answers):
    together += f"# Response from Model {index+1}\n\n{answer}\n\n"

display(Markdown(together))
```

```python
comparision_promp = f"""
You are an judge. Given the following responses from different language models to the same question, evaluate their quality based on accuracy, relevance, coherence, and informativeness. Provide a summary of strengths and weaknesses for each response, and conclude with a ranking from best to worst.
Question: {question}

Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks.

"""
```

```python
print(comparision_promp)
```

```python
judge = OpenAI()
evaluation = judge.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": comparision_promp}
    ]
)  
result = evaluation.choices[0].message.content
print(result)
```

```python
import json

ranked_results = json.loads(result)
display(Markdown("# Ranking of models from best to worst:"))
for rank, competitor in enumerate(ranked_results["results"], start=1):
    model_index = int(competitor) - 1
    model_name = models[model_index]
    display(Markdown(f"## {rank}. Model: {model_name} \n\n Response: {answers[model_index]}\n"))
```
