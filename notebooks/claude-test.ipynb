```python
import anthropic
import os
```

```python
client = anthropic.Anthropic(
    api_key=os.getenv("ANTHROPIC_API_KEY")
)
message = client.messages.create(
    model="claude-sonnet-4-5-20250929",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "What is the capital of France?"}
    ]
)
print(message.content[0].text)
```
